```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
```

```python

data = pd.read_csv('realtor-data-for-modeling.csv')
```

```python
data.head()
```

```python
data = data.drop(columns='Unnamed: 0')
```

```python
y = data.price

X = data.drop(columns="price")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
X_train.columns
```

```python
cols_to_use = ['bed','bath','acre_lot','house_size','sold_flag','log_acre','apt','house','Connecticut',
       'Maine', 'Massachusetts', 'New Hampshire', 'New Jersey', 'New York','Rhode Island', 'Vermont']
```

```python
X_train[cols_to_use]
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train[cols_to_use] , y_train)
```

```python
lr.score(X_train[cols_to_use] , y_train)
```

```python
y_pred = lr.predict( X_train[cols_to_use] )
```

```python
RMSE = (np.mean( (y_train - y_pred)**2 ))**(1/2)
```

```python
RMSE
```

```python
y_test_pred = lr.predict( X_test[cols_to_use] )
```

```python
(np.mean( (y_test - y_test_pred)**2 ))**(1/2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
# Train the model on training data
# Takes a few seconds
rf.fit( X_train[cols_to_use], y_train);
```

```python
predictions = rf.predict( X_train[cols_to_use] )
```

```python
(np.mean( (y_train - predictions)**2 ))**(1/2)
```

```python
test_predictions = rf.predict( X_test[cols_to_use] )
```

```python
(np.mean( (y_test - test_predictions)**2 ))**(1/2)
```
