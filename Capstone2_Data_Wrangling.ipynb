```python
### Now let's try to put all that into a function..

def get_df_from_URL(URL):
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    tablestring = soup.find_all('pre')[0].text
    listofstrings = tablestring.splitlines()
    listoflists = [x.split(';') for x in listofstrings]
    df = pd.DataFrame(listoflists[1:], columns=listoflists[0])
    return df
    
```

```python

for week in range(2,15):
    #print(week)
    URL = 'http://rotoguru1.com/cgi-bin/fyday.pl?week=' + str(week) + '&game=fd&scsv=1'
    loopdf = get_df_from_URL(URL)
    #df.concat(loopdf, ignore_index=True)
    df = pd.concat([df,loopdf] , ignore_index=True )
    
```
